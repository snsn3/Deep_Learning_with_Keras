```python
#imports - necessary libs
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
import warnings
warnings.simplefilter('ignore', FutureWarning)
import sklearn
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

I have grouped all imports here (as opposed to importing separately for each step or question)

```python
# this dataset can be downloaded from here: https://cocl.us/concrete_data
mes_donnees=pd.read_csv('/content/donnees_Coursera_10_concrete_data.csv')
mes_donnees.head() #ensuring correct data was loaded
```

# Data Source
https://cocl.us/concrete_data

```python
mes_donnees.shape #to check what we got
```

```python
mes_donnees.describe() #analyse descriptive
```

```python
mes_donnees_columns = mes_donnees.columns
pred = mes_donnees[mes_donnees_columns[mes_donnees_columns != 'Strength']] # since Strength is our tgt
tgt = mes_donnees['Strength'] # our target variable (predicted)
```

```python
pred.head() # to check our predictors - all looks good
```

```python
tgt.head() #to check our target - strength - all looks good
#Target is the strength of cement based on the volume of inputs (le contenu du ciment par volume)
```

```python
# I will calculate the mean and standard deviation manually
pred_mean = pred.sum(axis=0) / len(pred)
pred_var = ((pred - pred_mean) ** 2).sum(axis=0) / len(pred)
pred_std = pred_var ** 0.5
# Normalization of predictor vars
pred_norm = (pred - pred_mean) / pred_std
pred_norm = pd.DataFrame(pred_norm, columns=pred.columns)
pred_norm.head()
```

```python
n_cols = pred_norm.shape[1] # number of pred vars.
```

```python
# model
def regression_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
# model build
model = regression_model()
```

```python
# fit the model
model.fit(pred_norm, tgt, validation_split=0.3, epochs=50, verbose=2)
```

```python
# Randomly split the data: 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(pred_norm, tgt, test_size=0.3, random_state=42)
# Using X_train and y_train to fit model, and X_test, y_test for evaluation.
def evaluate_model(model, X_test, y_test):
    # Predict concrete strength on the test data
    predictions = model.predict(X_test)
    # Compute the mean squared error
    mse = mean_squared_error(y_test, predictions)
    return mse
mse = evaluate_model(model, X_test, y_test)
print("Mean Squared Error on test data:", mse)
```

```python
#list for storing the mean srd errors
mse_list=[]
#loop to repeat 50 times
for i in range(50):
  X_train, X_test, y_train, y_test = train_test_split(pred_norm, tgt, test_size=0.3, random_state=i)
  model.fit(X_train, y_train, validation_split=0.3, epochs=50, verbose=0)
  mse = evaluate_model(model, X_test, y_test)
  #computing the mse to the list
  mse_list.append(mse)
  #computing the mean & std of the MSEs
  mse_mean = np.mean(mse_list)
  mse_std = np.std(mse_list)
#print results
print("Mean of MSEs:", mse_mean)
print("Standard Deviation of MSEs:", mse_std)
```
